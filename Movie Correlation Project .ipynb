```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts the configuration of the plots made

#Read in the data

df = pd.read_csv(r'/Users/andrewgale/Desktop/movies.csv')
```

```python
# Looking at the data

df.head()
```

```python
# Check For Missing Data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Remove Any Null Values 

df = df.dropna(how='any',axis=0)

# Show Data types for each column

df.dtypes
```

```python
# Change Datatype of columns 

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df
```

```python
# Create 'Correct Year' Column

df['yearcorrect'] = df['released'].astype(str)
df['yearcorrect'] = df['yearcorrect'].str.rsplit(',').str[-1] 
df
```

```python
#Sorting By 'Gross' in Descending Order

df = df.sort_values(by=['gross'], inplace=False, ascending=False) 
```

```python
# Set view option to entire dataframe with following code pd.set_option('display.max_rows', None)
```

```python
# If Necessary, Drop Any Duplicates

df['company'].drop_duplicates().sort_values(ascending = False)

# To drop duplicates in the entire dataframe just type df.drop_duplicates()
```

```python
# Hypothesis 1 - Budget Is High Correlation with Gross

# Hypothesis 2 - Company High Correlation with Gross


#Make Scatter Plot with Budget vs. Gross Revenue

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs. Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show
```

```python
df.head()
```

```python
# Plot the Budget Vs. Gross Using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
#Start Looking at Correlation 

df.corr(method='pearson') #Pearson is default. There is also Kendall, and Spearman
```

```python
# There is high correlation between budget and gross

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
# Looking at Company Now - Make 'Object' a 'Numeric' Value in Order to Perform Correlation Operation

df.head()
```

```python
df_numerized = df 

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        

df_numerized
```

```python
#Compare Original Table as Objects, to Table with Numeric Values in Place of Objects

df 
```

```python
# Create A New and More Detailed Correlation Chart

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
df_numerized.corr()
```

```python
# View the Correlations in an easy-to-view Ordered List 

correlation_matrix = df_numerized.corr()

corr_pairs = correlation_matrix.unstack()

pd.set_option('display.max_rows', None)

corr_pairs
```

```python
# View Each Individual Pair of Correlations

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Only Using High Correlation Data Now

high_corr = sorted_pairs[(sorted_pairs > 0.5)]

high_corr
```
